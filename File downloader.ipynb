```python
import requests
import shutil
from tqdm.auto import tqdm
import urllib.parse as up
from pathlib import Path
import os

def get_filename_from_url(url):
    from urllib.parse import urlparse
    parsed_url = up.urlparse(url)
    
    try:
        filename = parsed_url.path
    except:
        # If it fails it is because the parsed_url implementation does not recognize attributes like "path"
        if len(parsed_url) >= 4:
            filename = parsed_url[2]
        else:
            filename = ""

    if "/" in filename:
        filename = filename.split("/")[-1]
    print(filename)
    return filename 

url=input("Enter URL to continue...")
fname2=get_filename_from_url(url)

def download(url):
    with requests.get(url, stream=True) as r:
    
    # check header to get content length, in bytes
        total_length = int(r.headers.get("Content-Length"))
        
        # implement progress bar via tqdm
        with tqdm.wrapattr(r.raw, "read", total=total_length, desc="")as raw:
            
            dPath = '/content/'+fname2
            # save the output to a file
            with open(f"{dPath}", 'wb')as output:
                shutil.copyfileobj(raw, output)
                print('Download completed!')
                
def bar_custom(current, total, width=80):
    print("Downloading: %d%% [%d / %d] bytes" % (current / total * 100, current, total))
```
